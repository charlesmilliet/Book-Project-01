```python
import pandas as pd #similar to dataframe in R 

#load data from desktop location
data=pd.read_csv(r"C:\Users\caiji\OneDrive\Desktop\Python Group Project\NYT_bestseller.csv")

#add the year column to the data set and the month-year column to the data set
data['month_year'] = pd.to_datetime(data['published_date']).dt.to_period('M')
data['year'] = pd.to_datetime(data['published_date']).dt.to_period('Y')

```

```python
#YEAR for function 1

year_start=input("Input start year :  ")
year_end=input("Input end year :  ")

if year_end>=year_start:
    print("yes")
    filtered=data[data["year"]>=year_start]
    filtered=filtered[filtered["year"]<=year_end]
else:
    print("The end date is before the start date")

filtered.head()
#filered["title”]
```

```python
#MONTH-YEAR for function 2
month_year_start=input("Input start year/month as YYYY-MM :  ")
month_year_end=input("Input end year/month as YYYY-MM :  ")

if month_year_end>=month_year_start:
    print("yes")
    filtered=data[data["month_year"]>=month_year_start]
    filtered=filtered[filtered["month_year"]<=month_year_end]
    print(filtered)

else:
    print("The end date is before the start date")
```
